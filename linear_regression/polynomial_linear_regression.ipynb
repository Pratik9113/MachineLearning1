```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data_set = pd.read_csv("E:\program\machine learning\linear_regression\Dataset_master.xlsx - RBC in Humans Polynomial Regres.csv")
X = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values
```

```python
from sklearn.linear_model import  LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X,y)
plt.scatter(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
linear_reg_2 = LinearRegression()
linear_reg_2.fit(X_poly,y)
```

```python
#visulize the linear regressor
plt.scatter(X,y,color='red')
plt.plot(X,linear_reg.predict(X), color='blue')
```

```python
# for polynomial result 
plt.scatter(X,y,color='red')
plt.plot(X,linear_reg_2.predict(X_poly), color='blue')
```
