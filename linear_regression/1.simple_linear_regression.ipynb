```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data_set = pd.read_csv("E:\program\machine learning\linear_regression\Dataset_master.xlsx - Simple Linear Regression (Salar.csv")
X = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values
```

```python
print(X)
print(y)
```

```python
# we split into train and test 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#training and test the data set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#prediction on test set
y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train), color='blue')
plt.title("ctc vs year of experience")
plt.xlabel("in year")
plt.ylabel("ctc")
plt.show()
```
