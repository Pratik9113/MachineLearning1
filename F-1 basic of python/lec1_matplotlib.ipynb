```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
x1 = [1,2,3,4,5]
y = [1,2,3,4,5]
y1 =[5,4,3,2,1]
plt.scatter(x,y)
plt.scatter(x1,y1)
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
# changing the color
x = [1,2,3,4,5]
x1 = [1,2,3,4,5]
y = [1,2,3,4,5]
y1 =[5,4,3,2,1]
plt.scatter(x1,y1)
plt.plot(x,y,"g-.")
plt.plot(x1,y1)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = 2*(x**4) +100
plt.plot(x,y)
plt.show()
```

```python
# arange is a function which help in smooth the curve

x = np.arange(0,10,0.0001)
y = 2*(x**4) +100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4
plt.plot(x,y, color ="red", linewidth =0.5, marker = "p")
plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4
y2 = x**3
plt.plot(x,y, color ="red", linewidth =0.5, label = "x^4")
plt.plot(x,y2, label = "x^3")
plt.ylabel("Y", fontsize ="15")
plt.xlabel("X" , fontsize ="15")
plt.legend()
plt.title("One Man Army")
plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4
y2 = x**3
plt.plot(x,y, color ="red", linewidth =0.5, label = "x^4")
plt.plot(x,y2, label = "x^3")
plt.ylabel("Y", fontsize ="15")
plt.xlabel("X" , fontsize ="15")
plt.legend()
plt.title("One Man Army")
plt.axis([0,15,0,10000])
plt.text(8,6000, "*" , fontsize = "15")
plt.grid()
plt.show()
```
