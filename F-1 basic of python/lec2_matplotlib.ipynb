```python
import matplotlib.pyplot as plt
import numpy as np
profit = [10,20,30,40]
value = ["Amazon", "Flipkart" ,"Oracle", "Apple"]
plt.pie(profit)
```

```python
profit = [10,20,30,49]
value = ["Amazon", "Flipkart" ,"Oracle", "Apple"]
col = ["blue" , "pink" ,"grey" ,"red"]
plt.pie(profit, colors= col , labels=value, autopct= "%.2f%%" )
```

```python
profit = [10,20,30,49]
value = ["Amazon", "Flipkart" ,"Oracle", "Apple"]
col = ["blue" , "pink" ,"grey" ,"red"]
space = [0.05 , 0.1, 0.0 , 0.0]
plt.pie(profit, colors= col , labels=value, autopct= "%.2f%%" , explode = space  )
```

```python
profit = [10,20,30,49]
value = ["Amazon", "Flipkart" ,"Oracle", "Apple"]
col = ["blue" , "pink" ,"grey" ,"red"]
space = [0.05 , 0.1, 0.0 , 0.0]
plt.pie(profit, colors= col , labels=value, autopct= "%.2f%%" , explode = space , counterclock=False, startangle=90 )
```
