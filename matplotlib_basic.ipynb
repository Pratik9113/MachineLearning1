```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sinx")
plt.title("sinwave")
plt.show()
```

```python
plt.plot(x,z)
plt.show()

```

```python
#parabola
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y , 'r+')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x) , 'g+')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
fig = plt.figure()
#create empty plot
ax = fig.add_axes([0,0,1,1])
#index reprent area
languages = [ 'english', 'hindi' , 'bangali', 'marathi' , 'france']
people = [100,30,244,24,242]
ax.bar(languages,people)
plt.xlabel('lan')
plt.ylabel('people')
plt.show()
```

```python
#pie chart t
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = [ 'english', 'hindi' , 'bangali', 'marathi' , 'france']
people = [100,130,244,156,242]
ax.pie(people,labels =languages, autopct = '%1.1f%%')
plt.show()
```

```python
#scatter plot
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color = 'b')
plt.show()
```

```python
#3d scatter plot
fig3 = plt.figure()
ax = plt.axes(projection ='3d')
z = 20 * np.linspace(-10,10,100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
